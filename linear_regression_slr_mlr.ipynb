```python
# Divide the data into independent vs dependent variables

# x = everything except profit
# y = profit

x = df.drop('Profit', axis=1).values
y = df['Profit'].values
```

```python
# Convert categorical data using Feature Encoding

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# the [3] is the categorical column
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
x_enc = np.array(ct.fit_transform(x))
x_enc
```
