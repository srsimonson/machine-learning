# Mean Shift Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('../../data_sets/mall.csv')
```

```python
dataset.head()
```

```python
X = dataset.iloc[:, [3,4]].values
```

## Fit mean shift cluster to dataset

```python
from sklearn.cluster import MeanShift
```

```python
ms = MeanShift(bandwidth=20)
# adjust the bandwith parameter to influence the number of clusters
```

```python
ms.fit(X)
labels = ms.labels_
cluster_centers = ms.cluster_centers_
```

```python
# Visualize the clusters

n_clusters = len(np.unique(labels))
```

```python
labels
# 7 clusters are formed
```

```python
colors = ['r', 'b', 'g', 'y', 'c', 'purple', 'm']

for i in range(n_clusters):
    cluster = X[labels == i]
    plt.scatter(cluster[:,0], cluster[:,1], s=100, c=colors[i], label=f'Cluster {i+1}')
plt.title('Cluster of Clients')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score (0-100)')
plt.legend()
plt.show()
```
