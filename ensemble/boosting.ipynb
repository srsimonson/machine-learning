```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('../../data_sets/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges, errors='coerce')
```

```python
df.dropna(how='any', inplace=True)
```

```python
X = df.drop(['customerID', 'Churn'], axis=1)
y = df.Churn.values
```

```python
X = pd.get_dummies(X, columns=['gender', 'Partner', 'Dependents','PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod'], drop_first=True)
```

```python
X.head()
```

```python
bool_cols = X.select_dtypes(include='bool').columns
X[bool_cols] = X[bool_cols].astype(int)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

## Ensemble Learning

### Adaboost

```python
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

```python
model_ada = AdaBoostClassifier(n_estimators=200)

model_ada.fit(X_train_sc, y_train)

y_pred_ada = model_ada.predict(X_test_sc)

print('Accuracy of Adaboost: ', accuracy_score(y_test, y_pred_ada))
```

### Gradient boost

```python
model_gb = GradientBoostingClassifier(n_estimators=200)

model_gb.fit(X_train_sc, y_train)

y_pred_gb = model_gb.predict(X_test_sc)

print('Accuracy of GradientBoost: ', accuracy_score(y_test, y_pred_gb))
```
