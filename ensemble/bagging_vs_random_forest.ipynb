# Bagging vs Random Forest

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.tree import plot_tree
```

```python
X, y = make_classification(n_samples=200, n_features=5, n_redundant=0, n_informative=5, n_clusters_per_class=1)
```

```python
df = pd.DataFrame(X, columns=['Col 1','Col 2','Col 3','Col4','Col 5'])
df['target']=y
```

```python
df.head()
```

```python
len(df)
```

```python
df.shape
# rows, columns
```

```python
bag = BaggingClassifier(max_features=2)
```

```python
bag.fit(df.iloc[:,:5], df.iloc[:,-1])
```

```python
plt.figure(figsize=(10,10))
plot_tree(bag.estimators_[0])
plt.show()
```

```python
rf = RandomForestClassifier(max_features=2)
rf.fit(df.iloc[:,:5], df.iloc[:,-1])
```

```python
plt.figure(figsize=(10,10))
plot_tree(rf.estimators_[0])
plt.show()
```

```python
# See that Random Forest is much more comprehensive. Bagging always split just x[0] and x[1] whereas
# Random Forest randomizes many more variables.
```
